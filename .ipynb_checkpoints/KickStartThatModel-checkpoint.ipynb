```python
import time
import pandas as pd
import json
import datetime
import numpy
```

```python
tp_df = pd.read_csv("data/KS_data.csv").dropna()
tp_df[tp_df["ks_state"] == "live"].head()
tp_df.head()
```

```python
tp_df = tp_df.drop(['Unnamed: 0'], axis=1)
tp_df = tp_df.drop(['game_id'], axis=1)
tp_df = tp_df.drop(['location_id'], axis=1)
tp_df.head()
```

```python
tp_df["ks_state"][tp_df["ks_state"] == "successful"] = 1
tp_df["ks_state"][tp_df["ks_state"] != 1] = 0
ks_money_list = []
tp_df["ks_money_made"] = [ "<100%" if p < g else\
                         "100%-125%" if p <= g*1.25 else\
                         "125%-150%" if p <= g*1.5 else\
                         "150%-200%" if p <= g*2 else\
                         "200%-500%" if p <= g*5 \
                          "500%-1000%" if p <= g*10 \
                          else ">1000%"
                          for p,g in zip(tp_df["money_pledged"], tp_df["money_goal"])]



X = tp_df.drop(["ks_state"], axis=1).drop(["money_pledged"], axis=1).drop(["ks_money_made"], axis = 1)
# X= tp_df[["backers_count","ks_state","backers_count","money_goal"]]
y = tp_df[["ks_money_made"]]
print(X.shape, y.shape)
X.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from keras.utils import to_categorical
import seaborn as sns
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1)
sns.pairplot(tp_df[["money_pledged",
               "backers_count",
               "money_goal",
                "ks_money_made"]], diag_kind="kde")
```

```python
# y_scaler = MinMaxScaler().fit(y_train)
# y_train_scaled = y_scaler.transform(y_train)
# y_test_scaled = y_scaler.transform(y_test)
# y_test_scaled[:5]
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_encoder.fit(y_train)
# print(y_train)
encoded_y_train = label_encoder.transform(y_train)
print(encoded_y_train[0])
y_train_scaled = to_categorical(encoded_y_train)
print(y_train_scaled[0])

y_train_scaled
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_test_scaled[:5]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

model = Sequential()
model.add(Dense(100, input_dim=X_test_scaled.shape[1], activation='relu'))
model.add(Dense(80, activation='relu'))
model.add(Dense(6, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer="adam",metrics=['mean_absolute_error', 'mean_squared_error'])
```

```python
model.summary()
```

```python
model.fit(X_train_scaled, y_train_scaled, epochs=20, verbose=2)
# new instances where we do not know the answer
```
